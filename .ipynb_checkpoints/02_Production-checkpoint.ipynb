```python
bear_types = 'grizzly', 'black', 'teddy'
path = Path('bears')
# results = search_images_bing(key, f'teddy bear')

doc(path.mkdir)
doc(download_images)
if not path.exists():
  path.mkdir()
  for t in bear_types:
    dest = (path/t)
    dest.mkdir(exist_ok=True)
    print(f'made dir: {dest}')
    results = search_images_bing(key, f'{t} bear')
    download_images(dest, urls=results.attrgot('contentUrl'))
    print(f'downloaded {t} bear images: {len(results)}')
```

```python
import os
grizzly_path = 'bears/grizzly'
file_name = os.listdir(grizzly_path)[2]
full_path = f'{grizzly_path}/{file_name}'
Image.open(full_path).to_thumb(128, 128)
```

```python
bears = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
path = Path('bears')
print(f'path: {path}')
dls = bears.dataloaders(path)
type(dls)
```

```python
# dir(len), dir(len[0])
type(dls), len(dls), len(dls[0].items), len(dls[1].items), dls[0].items[0]
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
bears = bears.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = bears.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

```python
# dir(dls)
import inspect
attributes = inspect.getmembers(DataLoaders, lambda m: not(inspect.isroutine(m)))
[a for a in attributes if a[0].startswith('_docs')]
[a for a in attributes if not((a[0].startswith('_')) or (a[0].startswith('__') and a[0].endswith('__')))]
```

```python
doc(aug_transforms)
bears = bears.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls = bears.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
bears = bears.new(item_tfms=Resize(224), batch_tfms=None)
dls = bears.dataloaders(path)
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
interp.plot_top_losses(5, nrows=1)
# TODO: How to map the top losses to corresponding image file path
```

```python
bears = bears.new(item_tfms=RandomResizedCrop(224, min_scale=0.5))
dls = bears.dataloaders(path)
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
interp.plot_top_losses(5, nrows=1)
# TODO: How to map the top losses to corresponding image file path
```

```python
bears = bears.new(item_tfms=RandomResizedCrop(224, min_scale=0.5), batch_tfms=aug_transforms())
dls = bears.dataloaders(path)
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
interp.plot_top_losses(5, nrows=1)
# TODO: How to map the top losses to corresponding image file path
```
